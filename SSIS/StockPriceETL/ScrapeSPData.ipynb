```python
import yfinance as yf
```

##### Download Microsoft Stock price from Nov 2022 to Nov 2025

```python
Microsoft = yf.download(
            tickers="MSFT", start="2022-11-01", end="2025-11-01", 
            interval="1d", multi_level_index=False)
Microsoft.reset_index(inplace=True) 
Microsoft.index = range(1, len(Microsoft) + 1)
```

```python
Microsoft.head()
# Microsoft[Microsoft['Date'].dt.year == 2025]
```
